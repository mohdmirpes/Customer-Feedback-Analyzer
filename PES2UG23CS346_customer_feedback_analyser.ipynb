```python
from transformers import pipeline
```

```python
sentiment_analyzer = pipeline("sentiment-analysis")
```

```python
def analyze_feedback(text):
    result = sentiment_analyzer(text)
    label = result[0]["label"]
    score = result[0]["score"]
    return label, score
```

```python
review = "The product quality is excellent and I am very satisfied."

label, score = analyze_feedback(review)

print("Review:")
print(review)

print("\nSentiment:", label)
print("Confidence Score:", round(score, 3))
```

```python
reviews = [
    "The service was amazing and the staff was very helpful.",
    "I am disappointed with the product quality.",
    "Delivery was late and customer support was unresponsive.",
    "Great value for money. Highly recommend!",
    "The experience was average, nothing special."
]

for r in reviews:
    label, score = analyze_feedback(r)
    print(f"Review: {r}")
    print(f"Sentiment: {label}, Confidence: {round(score, 3)}\n")
```

```python
positive = 0
negative = 0

for r in reviews:
    label, _ = analyze_feedback(r)
    if label == "POSITIVE":
        positive += 1
    else:
        negative += 1

print("Summary:")
print("Positive Reviews:", positive)
print("Negative Reviews:", negative)
```

```python
user_review = input("Enter a customer review: ")

label, score = analyze_feedback(user_review)

print("\nSentiment:", label)
print("Confidence Score:", round(score, 3))
```
